## Tratando os Dados

```python
data_train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
data_test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
data_train.isna().sum().to_frame()
```

```python
data_train['Age'].fillna(data_train['Age'].median(), inplace=True)

data_train.dropna(axis=0, inplace=True)
```

```python
data_test.isna().sum().to_frame()
```

```python
data_test['Age'].fillna(data_test['Age'].median(), inplace=True)

data_test['Fare'].fillna(data_test['Fare'].median(), inplace=True)
```

## Criação do Modelo de ML

```python
from sklearn.preprocessing import LabelEncoder

# Utilizando o LabelEncoder

data_train['Sex'] = data_train[['Sex']].apply(LabelEncoder().fit_transform)
data_train['Embarked'] = data_train[['Embarked']].apply(LabelEncoder().fit_transform)

data_test['Sex'] = data_test[['Sex']].apply(LabelEncoder().fit_transform)
data_test['Embarked'] = data_test[['Embarked']].apply(LabelEncoder().fit_transform)

data_train.head()
```

```python
data = data_train.drop(['Survived'], axis=1)
target = data_train['Survived']
```

### Usando Logistic Regression

```python
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(data, target)
```

```python
score_lr = round(lr_model.score(data, target) * 100, 2)

score_lr
```

```python
y_pred_lr = lr_model.predict(data_test)

data_lr = data_test.copy()

data_lr['Prediction'] = y_pred_lr
data_lr
```

```python
submission = pd.DataFrame({"PassengerId": passengerId,
                           "Survived": y_pred_lr})

# Gerando o arquivo

submission.to_csv('submission_lr.csv', index=False)
```

### Usando Random Forest

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
rf_model.fit(data, target)
```

```python
score_rf = round(rf_model.score(data, target) * 100, 2)

score_rf
```

```python
y_pred_rf = rf_model.predict(data_test)

data_rf = data_test.copy()

data_rf['Prediction'] = y_pred_rf
data_rf
```

```python
submission = pd.DataFrame({"PassengerId": passengerId,
                           "Survived": y_pred_rf})

# Gerando o arquivo

submission.to_csv('submission_rf.csv', index=False)
```
